```python
import gradio as gr
import tensorflow as tf
```

```python
from transformers import TFGPT2LMHeadModel, GPT2Tokenizer
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = TFGPT2LMHeadModel.from_pretrained("gpt2", pad_token_id=tokenizer.eos_token_id)
```

```python
inp="I love ice creaam"
input_ids = tokenizer.encode(inp, return_tensors='tf')
beam_output = model.generate(input_ids, max_length=100, num_beams=5, no_repeat_ngram_size=2, early_stopping=True)
# beam_output
print(tokenizer.decode(beam_output[0], skip_special_tokens=True, clean_up_tokenization_space=True))
```

```python
def generate_blog(inp):
  input_ids = tokenizer.encode(inp, return_tensors='tf')
  beam_output = model.generate(input_ids, max_length=1000, num_beams=5, no_repeat_ngram_size=2, early_stopping=True)
  output = tokenizer.decode(beam_output[0], skip_special_tokens=True, clean_up_tokenization_space=True)
  return ".".join(output.split(",")[:-1]) + '.'
```

```python
output_text = gr.outputs.Textbox()

gr.Interface(generate_blog, "textbox", output_text, title="Blog generation deeplearning").launch()
```
